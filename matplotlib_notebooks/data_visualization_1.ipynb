```python
import matplotlib.pyplot as plt
```

```python
age=[25,27,29,31,33,35,37,39,41,43,45] # list(range(25,46,2))


salary=[38476, 42000, 46752,49320,53200,56000,62316,64928,67317,68748,73752]
```

```python
plt.plot(age,salary)
plt.xlabel('age')
plt.ylabel('salary')
plt.title('Salary by Age')
plt.show()
```

```python
salary2=[45372,48876,53850,57287,63016,65998,70003,70000,71496,75370,83640]
```

```python
plt.plot(age,salary, label='salary')
plt.plot(age,salary2, label='salary2')

plt.xlabel('age')
plt.ylabel('salary')
plt.title('Salary by Age')
plt.legend()
plt.show()
```

```python
#subplot
plt.subplot(2,1,1)# 2 row, 1 column, I am drawing the first one
plt.plot(age,salary, 'r')# last parameter is color (red here)
plt.subplot(2,1,2)# 2 row, 1 column, I am drawing the second one
plt.plot(age,salary, 'b')
plt.tight_layout()
```

```python
plt.subplot(2,2,1)
plt.plot(age,salary, 'r')# last parameter is color (red here)
plt.subplot(2,2,2)# 2 row, 1 column, I am drawing the second one
plt.plot(age,salary, 'b')
plt.subplot(2,2,3)# 2 row, 1 column, I am drawing the second one
plt.plot(age,salary, 'g')

plt.tight_layout()
```

```python
# We can take plot with one variable set
plt.plot(age)
#Here, x axis is the count of values, y axis is the age
```

```python
import pandas as pd 
df=pd.DataFrame(list(zip(age,salary,salary2)), columns=['age', 'salary', 'salary2'])
df.head()
```

```python
plt.plot(df['age'],df['salary']);
```

```python
#OOP method (Manual)
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])#Those numbers are to define the proportional area for the plot (x,y,width,height)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(age,salary, 'r', label='salary by age')
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_title('Salary by age')
plt.legend();
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([1.1,0.1,0.4,0.4])


ax1.plot(age,salary, 'r')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.set_title('Salary by age')

ax2.plot(age,salary2, 'r')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary2')
ax2.set_title('Salary2 by age')
```

```python
#adding axes automatically
fig,ax=plt.subplots()
ax.plot(age,salary, 'r')
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_title('Salary by age')
plt.tight_layout()
```
