```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
age=[25,27,29,31,33,35,37,39,41,43,45]
salary=[38496,42000,46752,49320,53200,56000,62316,64928,67317,68748,73752]
salary_2=[45372,48876,53850,57287,63016,65998,70003,70000,71496,75370,83640]
```

```python
fig=plt.figure(figsize=(8,4))#axes te bir ekleme degisiklik vs yapilmayacaksa boyle olusturulabilir.
#Buradaki (8,4) inch cinsinden ebatlar, vermezsek default (6.4,4.8)
```

```python
fig,ax=plt.subplots(figsize=(6,3))#Bu yontem her turlu calisir, daha cok bunu kullanacagiz.
ax.plot(age,salary,'r')
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.set_title('Ages with Salary')
```

```python
fig.savefig('myplot.png', dpi=100)#jpeg veya pdf olarak da kaydedilebilir.
```

```python
fig,ax=plt.subplots(figsize=(8,3))#Bu yontem her turlu calisir, daha cok bunu kullanacagiz.
ax.plot(age,salary,'r', label='salary_1', lw=4, ls='-.', marker='o', ms=10, markerfacecolor='b', alpha=0.3)
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.set_title('Ages with Salary')
ax.legend()
```

```python
#PLOT RANGE
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(age,salary)
ax.set_xlim([30,40])#Belli bir alana odaklanmak icin xlim ve ylim atayabiliyoruz.
ax.set_ylim([45000,60000])
#Adding vertical-horizantal lines
ax.axvline(x=35, color='green', ls='--')
ax.axhline(y=52000, color='orange')
```

```python
# Bar Plot
country=['UK','USA','FRA', 'GER', 'NOR']
pci=[40000,50000,38000,55000,80000]
```

```python
fig,ax=plt.subplots()
ax.bar(country, pci)
ax.set_xlabel('country')
ax.set_ylabel('pci')
ax.set_title('pci per country')
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```python
fig,ax=plt.subplots()
ax.bar(labels,men_means);
```

```python
fig,ax=plt.subplots()
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

ax.bar(x - width/2, men_means, width, label='Men')# x - width/2 bu kisim ile x eksenindeki yerini ayarliyoruz.
#Yoksa ust uste biner iki plot
ax.bar(x + width/2, women_means, width, label='Women')
```
