```python
import numpy as np

np.random.seed(101)

population=np.random.randint(1,80,100000)

population.mean()

```

```python

np.random.seed(101)

sample_means=[]

for i in range(10):

    sample=np.random.choice(population,100)

    sample_means.append(sample.mean())

np.mean(sample_means)

```

```python
# skewness and kurtosis
from scipy.stats import kurtosis,skew

import matplotlib.pyplot as plt
x=np.random.normal(0,2,100000)
```

```python
plt.hist(x,bins=100)
```

```python
kurtosis(x)
```

```python
skew(x)
```

```python
shape,scale=2,2
s=np.random.gamma(shape,scale,100000)
```

```python
plt.hist(s,bins=100)
```

```python
skew(s)
```

```python
kurtosis(s)
```
