```python
from scipy.stats import kurtosis,skew
from scipy import stats
import matplotlib.pyplot as plt 
```

```python
x=np.random.normal(0,2,1000000)#ortalamasi 0, standart sapmasi 2 olan 1000000 elemanli bir normal dagilim olusturduk.
```

```python
plt.hist(x,bins=100);
```

```python
kurtosis(x)#normal dagilimin kurtosis i 3 e yakin bir degerdir. Python la hesaplatinca 3 cikararak sonucu verir. Yani 0 a yakin bir sonuc doner.
#kurtosis ten 3 cikinca kalan degere excess kurtosis denir.
#Finans dunyasi icin yuksek excess kurtosis yuksek risk gostergesidir.
```

```python
skew(x)
#normal dagilimin carpikligi 0 a yakindir.
```
